# Analysis and Visualisation

## Rating of Movies and Shows

```python
sns.countplot(dataset_copy['rating'])
fig = plt.gcf()
fig.set_size_inches(13,13)
```

## Relationship b/w type and rating of movies and shows 

```python
sns.countplot(x='rating',hue='type',data=dataset_copy)
fig = plt.gcf()
fig.set_size_inches(13,13)
```

## Relation between Movies and TV Shows

```python
sns.countplot(dataset_copy['type'])
fig = plt.gcf()
fig.set_size_inches(13,13)
```

## Which country has the most number of titles produced?

```python
most_titles_produced = dataset_copy.groupby('country').count().sort_values('title',ascending=False).head(5)

most_titles_produced.reset_index(inplace=True)
most_titles_produced
```

```python
plt.bar(most_titles_produced.country,most_titles_produced['title'])
```
